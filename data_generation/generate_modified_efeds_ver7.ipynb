```python
from astropy.table import QTable
import mceh.find_luminosity as fl # v3.0.0-alpha2
from copy import deepcopy
from importlib import reload
import numpy as np
```

```python
efeds = QTable.read('data/modified_efeds_ver6.fits')
```

```python
new_efeds = deepcopy(efeds)
```

```python
efeds_band_name = ['g_cmag', 'r_cmag', 'i_cmag', 'z_cmag', 'y_cmag']
band_name = ['hsc_g', 'hsc_r', 'hsc_i', 'hsc_z', 'hsc_y']
for c_band, band in zip(efeds_band_name, band_name):
    new_efeds[c_band] = fl.cmag(new_efeds['Z_BEST_COMB'], band)
```

```python
for c1, c2 in zip(efeds, new_efeds):
    print(c1['i_cmag'], c2['i_cmag'])
```

```python
new_efeds['unmasked_fraction'] = new_efeds['unmasked_area'] / new_efeds['area']
```

```python
new_efeds.write('data/modified_efeds_ver7.fits')
```
