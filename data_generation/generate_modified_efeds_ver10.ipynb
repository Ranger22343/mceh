```python
from astropy.table import QTable
from astropy.io import fits
from scipy.interpolate import interp1d
import numpy as np
import mceh.find_luminosity as fl
import tqdm
from importlib import reload
import astropy.units as u
```

```python
original_efeds = QTable.read('data/modified_efeds_ver9.fits')
hsc = QTable.read('data/modified_hsc_ver5.fits')
rd = QTable.read('data/modified_random_ver3.fits')
```

```python
original_efeds.colnames
```

```python
# The columns that need to be changed is unmasked_area, grizy_cmag, galaxy_index, unmasked_fraction
```

```python
galaxy_index = []
for i in range(len(original_efeds)):
    galaxy_index.append(fl.index_within_angular_radius(hsc['ra'], hsc['dec'], 
                                   original_efeds['RA_CORR'][i], 
                                   original_efeds['DEC_CORR'][i], 
                                   original_efeds['r500c_lcdm'][i]))
    print(f'Finish {i+1}/{len(original_efeds)}')
```

```python
galaxy_index
```

```python
all_umarea = []
for i in tqdm.trange(len(original_efeds)):
    all_umarea.append(fl.unmasked_area(original_efeds['RA_CORR'][i], original_efeds['DEC_CORR'][i], original_efeds['r500c_lcdm'][i], rd))
```

```python
all_umarea = [i.value for i in all_umarea] * u.arcmin**2
```

```python
all_umarea
```

```python
unmasked_fraction = (all_umarea / original_efeds['area']).value
unmasked_fraction
```

```python
(unmasked_fraction > 0.6) == (original_efeds['unmasked_fraction'] > 0.6)
```

```python
new_efeds = original_efeds.copy()
new_efeds['unmasked_area'] = all_umarea
new_efeds['galaxy_index'] = galaxy_index
new_efeds['unmasked_fraction'] = unmasked_fraction
```

```python
new_efeds.write('data/modified_efeds_ver10.fits', overwrite=True)
```
