```python
data = fits.getdata('/array/users/inchiu/redSequenceModel/rsproducts/pdr3/bc03_rs_zfp3d0_tau_p0d4.fits', ext=-1)

def get_cmag(z, band_name):
    mag_func = interp1d(data['redshift'], data['p1d000_Lstar_' + band_name])
    return mag_func(z)


print(np.unique(get_cmag(original_efeds['Z_BEST_COMB'], 'hsc_g') == original_efeds['g_cmag']))
print(np.unique(get_cmag(original_efeds['Z_BEST_COMB'], 'hsc_r') == original_efeds['r_cmag']))
print(np.unique(get_cmag(original_efeds['Z_BEST_COMB'], 'hsc_i') == original_efeds['i_cmag']))
print(np.unique(get_cmag(original_efeds['Z_BEST_COMB'], 'hsc_z') == original_efeds['z_cmag']))
print(np.unique(get_cmag(original_efeds['Z_BEST_COMB'], 'hsc_y') == original_efeds['y_cmag']))
```

```python
galaxy_index = []
for i in range(len(original_efeds)):
    galaxy_index.append(fl.index_within_angular_radius(hsc['ra'], hsc['dec'], 
                                   original_efeds['RA_CORR'][i], 
                                   original_efeds['DEC_CORR'][i], 
                                   original_efeds['r500c_lcdm'][i]))
    print(f'Finish {i+1}/{len(original_efeds)}')
```

```python
galaxy_index
```

```python
rd = QTable.read('data/modified_random_ver2.fits')
```

```python
all_umarea = []
for i in tqdm.trange(len(original_efeds)):
    all_umarea.append(fl.unmasked_area(original_efeds['RA_CORR'][i], original_efeds['DEC_CORR'][i], original_efeds['r500c_lcdm'][i], rd))
```

```python
all_umarea = [i.value for i in all_umarea] * u.arcmin**2
```

```python
all_umarea
```

```python
unmasked_fraction = (all_umarea / original_efeds['area']).value
unmasked_fraction
```

```python
new_efeds = original_efeds.copy()
new_efeds['unmasked_area'] = all_umarea
new_efeds['galaxy_index'] = galaxy_index
new_efeds['unmasked_fraction'] = unmasked_fraction
```

```python
new_efeds.write('data/modified_efeds_ver9.fits', overwrite=True)
```
